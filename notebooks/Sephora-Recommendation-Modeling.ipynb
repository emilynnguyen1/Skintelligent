### Calculating Scores
---
- Popularity Score = Loves + Number of Reviews
- Rating Score = Converts the 1-5 star scale to 0-1
- Worth The Hype Score (Polarization) =
    - High std deviation = polarized → hype penalty
        - Consistent ratings → high score
        - Polarizing product → lower score
- Ingredient Compatibility Score considers ingredient_name, acne_risk, irritation_risk, dry_skin_risk, sensitivity_risk

**Popularity Score**

```python
# Popularity score based on loves_count and review_count, normalized to 0-1 range
product_stats['popularity_raw'] = (
    product_stats['loves_count'] + product_stats['review_count']
)

product_stats['popularity_score'] = np.log1p(product_stats['popularity_raw'])

product_stats['popularity_score'] = (
    (product_stats['popularity_score'] - product_stats['popularity_score'].min()) /
    (product_stats['popularity_score'].max() - product_stats['popularity_score'].min())
)
```

```python
product_stats.head()
```

**Rating Score**

```python
# Confidence weighting so that products with more reviews have more influence on the final score,
# but we use log to avoid giving too much weight to products with very high review counts
product_stats['confidence_weight'] = (
    np.log1p(product_stats['review_count']) /
    np.log1p(product_stats['review_count'].max())
)

# Adjusted rating score
product_stats['rating_score'] = (product_stats['avg_review_rating'] - 1) / 4
product_stats['adjusted_rating_score'] = (
    product_stats['rating_score'] * product_stats['confidence_weight']
)
```

**Worth The Hype Score**

```python
product_stats['rating_std'] = product_stats['rating_std'].fillna(0)

product_stats['hype_score'] = 1 / (1 + product_stats['rating_std'])
```

**Ingredient Compatibility Score**

The ingredient compatibility score is computed by comparing product ingredient lists with a user’s skin profile, ingredient preferences, and ingredient restrictions. Ingredients associated with acne, irritation, or sensitivity reduce the compatibility score, while preferred ingredients increase the score. The final compatibility score ranges from 0 to 1 and is used as a major component of the overall recommendation ranking.

```python
ingredient_risk = pd.read_csv('../data/ingredient_risk_table.csv')
ingredient_risk.head(10)
```

```python
def ingredient_compatibility_score(ingredient_string, user_profile, ingredient_risk):
    
    if pd.isna(ingredient_string):
        return 0.5
    
    ingredients = ingredient_string.lower()
    
    score = 1.0
    penalties = 0
    bonuses = 0
    
    # weighted risk system
    risk_weight = {
        "acne_risk": 0.2,
        "irritation_risk": 0.25,
        "dry_skin_risk": 0.15,
        "sensitivity_risk": 0.25
    }
    
    # --- Preferred ingredients ---
    for ing in user_profile.get("ingredient_preferences", []):
        if ing.lower() in ingredients:
            bonuses += 0.1
    
    # --- Avoid ingredients ---
    for ing in user_profile.get("ingredient_avoid", []):
        if ing.lower() in ingredients:
            penalties += 0.25
    
    # --- Risk-based penalties ---
    for _, row in ingredient_risk.iterrows():
        ing = row["ingredient_name"]
        
        if ing in ingredients:
            
            if user_profile.get("acne_prone", False):
                penalties += row["acne_risk"] * risk_weight["acne_risk"]
            
            if user_profile.get("skin_sensitivity") in ["High", "Very High"]:
                penalties += row["irritation_risk"] * risk_weight["irritation_risk"]
            
            if user_profile.get("dry_skin"):
                penalties += row["dry_skin_risk"] * risk_weight["dry_skin_risk"]
    
    # --- Fragrance allergy ---
    if user_profile.get("fragrance_allergy") and "fragrance" in ingredients:
        penalties += 0.3
    
    final_score = score + bonuses - penalties
    
    return max(min(final_score, 1), 0)
```

```python
# function to generate reason codes for ingredient compatibiility
def ingredient_reason_codes(ingredient_string, user_profile):
    
    reasons = []
    
    if pd.isna(ingredient_string):
        return reasons
    
    ingredients = ingredient_string.lower()
    
    for ing in user_profile["ingredient_preferences"]:
        if ing.lower() in ingredients:
            reasons.append(f"Contains preferred ingredient: {ing}")
    
    for ing in user_profile["ingredient_avoid"]:
        if ing.lower() in ingredients:
            reasons.append(f"Contains ingredient to avoid: {ing}")
    
    if "fragrance" in ingredients and user_profile["fragrance_allergy"]:
        reasons.append("Contains fragrance (allergy warning)")
    
    return reasons
```
